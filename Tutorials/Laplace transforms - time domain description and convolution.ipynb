```julia
using Plots
```

```julia
# Define the piecewise function
function x(t)
    if t < 0
        return 0
    elseif 0 <= t <= 1
        return 2 * t
    elseif 1 < t <= 2
        return 2 * (2 - t)
    else
        return 0
    end
end
```

```julia
# Generate a range of t values
t_values = 0:0.01:3

# Compute x(t) for each t
x_values = [x(t) for t in t_values]

# Plot the function
plot(t_values, x_values, label="x(t)", xlabel="t", ylabel="x(t)", linewidth=2)
```
