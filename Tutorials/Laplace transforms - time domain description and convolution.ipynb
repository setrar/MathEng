```julia
using Plots
```

```julia
# Define the piecewise function
function x(t)
    if t < 0
        return 0
    elseif 0 <= t <= 1
        return 2 * t
    elseif 1 < t <= 2
        return 2 * (2 - t)
    else
        return 0
    end
end
```

```julia
# Generate a range of t values
t_values = 0:0.01:3

# Compute x(t) for each t
x_values = [x(t) for t in t_values]

# Plot the function
plot(t_values, x_values, label="x(t)", xlabel="t", ylabel="x(t)", linewidth=2)
```

```julia
using Plots
using DSP

# Define the piecewise function for x(t)
function x(t)
    if 0 <= t <= 2
        return 4
    else
        return 0
    end
end

# Define the piecewise function for h(t)
function h(t)
    if 0 <= t <= 1
        return 2
    else
        return 0
    end
end

# Generate time range for x(t) and h(t)
t_values_x = -.5:0.01:5  # Time for x(t), extending to accommodate convolution result
t_values_h = -.5:0.01:5  # Time for h(t)

# Sample x(t) and h(t) at discrete time points
x_values = [x(t) for t in t_values_x]
h_values = [h(t) for t in t_values_h]

# Perform convolution (approximate continuous convolution by discrete sampling)
y_values = conv(x_values, h_values) * 0.01  # Multiply by sampling interval (0.01)

# Generate time axis for the convolution result
t_values_y = 0:0.01:(length(y_values)-1)*0.01

# Plot x(t), h(t), and the convolution result y(t)
plot(t_values_x, x_values, label="x(t)", xlabel="t", ylabel="Amplitude", linewidth=2)
plot!(t_values_h, h_values, label="h(t)", xlabel="t", ylabel="Amplitude", linewidth=2)
plot!(t_values_y, y_values, label="y(t) = x(t) * h(t)", xlabel="t", ylabel="Amplitude", linewidth=2)
```
