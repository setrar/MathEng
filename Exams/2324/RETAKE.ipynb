# A linear, time-invariant system has the following transfer function:


$H(s) = \frac{20s^2 + 2000s}{(s + 2)(s + 10)}$

(a) Derive an expression for $H(s)$ in the usual, normal form.

(b) Determine the frequency-invariant gain $K$ and the position of any poles and zeros.

(c) Sketch a Bode plot of the magnitude-frequency response.

(d) Sketch a Bode plot of the phase-frequency response.

## (a) Derive an expression for $H(s)$ in the usual, normal form.


The given transfer function is:

$H(s) = \frac{20s^2 + 2000s}{(s + 2)(s + 10)}$

### Step 1: Expand the Denominator
Expand the denominator:

$(s + 2)(s + 10) = s^2 + 10s + 2s + 20 = s^2 + 12s + 20.$

### Step 2: Divide by the Leading Coefficient of the Denominator
In the "normal form," the denominator should have its leading coefficient as 1. To achieve this, divide both the numerator and denominator by 1 (since the coefficient of $s^2$ in the denominator is already 1):

$H(s) = \frac{20s^2 + 2000s}{s^2 + 12s + 20}.$

### Step 3: Factor Out the Gain Term
Factor 20 from the numerator:

$H(s) = \frac{20(s^2 + 100s)}{s^2 + 12s + 20}.$

This is already in its normal form, where the denominator is monic (leading coefficient is 1), and the numerator is factored for clarity.

Thus, the final expression in normal form is:

$H(s) = 20 \cdot \frac{s^2 + 100s}{s^2 + 12s + 20}.$

## (b) Determine the frequency-invariant gain $K$ and the position of any poles and zeros.

The given transfer function is:

$H(s) = \frac{20s^2 + 2000s}{(s + 2)(s + 10)}$

### Step 1: Identify the Frequency-Invariant Gain $K$
The frequency-invariant gain $K$ (also called the high-frequency gain) is determined by the ratio of the leading coefficients of the numerator and denominator.

- The coefficient of $s^2$ in the numerator is **20**.
- The coefficient of $s^2$ in the denominator is **1**.

Thus, the high-frequency gain is:

$K = \frac{20}{1} = 20.$

### Step 2: Find the Poles
Poles are found by solving for $s$ in the denominator:

$(s + 2)(s + 10) = 0.$

Setting this equal to zero:

$s + 2 = 0 \quad \Rightarrow \quad s = -2.$

$s + 10 = 0 \quad \Rightarrow \quad s = -10.$

Thus, the poles are at:

$s = -2, \quad s = -10.$

### Step 3: Find the Zeros
Zeros are found by solving for $s$ in the numerator:

$20s^2 + 2000s = 0.$

Factor out $s$:

$20s(s + 100) = 0.$

Setting this equal to zero:

$s = 0, \quad s = -100.$

Thus, the zeros are at:

$s = 0, \quad s = -100.$

### Final Answer:
- **Frequency-invariant gain**: $K = 20$.
- **Poles**: $s = -2, s = -10$.
- **Zeros**: $s = 0, s = -100$.

## (c) Sketch a Bode plot of the magnitude-frequency response in Julia code

```julia
using ControlSystems, Plots

# Define the transfer function H(s) = (20s^2 + 2000s) / ((s + 2)(s + 10))
s = tf("s")
H = (20s^2 + 2000s) / ((s + 2) * (s + 10))

# Define frequency range (log scale)
 = exp10.(range(-1, stop=3, length=500))  # Frequency range from 0.1 to 1000 rad/s

# Compute magnitude response
mag, _, _ = bode(H, )  # Extract magnitude response (in absolute values)

# Convert magnitude to decibels (dB)
mag_db = 20 * log10.(vec(mag))  

# Plot magnitude response
plot(, mag_db, xscale=:log10, xlabel="Frequency (rad/s)", ylabel="Magnitude (dB)",
     title="Bode Magnitude Plot", legend=false, grid=true)
```

## (d) Sketch a Bode plot of the phase-frequency response in julia code

```julia
using ControlSystems, Plots

# Define the transfer function H(s) = (20s^2 + 2000s) / ((s + 2)(s + 10))
s = tf("s")
H = (20s^2 + 2000s) / ((s + 2) * (s + 10))

# Define frequency range (log scale)
 = exp10.(range(-1, stop=3, length=500))  # Frequency range from 0.1 to 1000 rad/s

# Compute frequency response
_, phase, _ = bode(H, )  # Extract phase response (in degrees)

# Convert to a 1D array if necessary
phase = vec(phase)  

# Plot phase response
plot(, phase, xscale=:log10, xlabel="Frequency (rad/s)", ylabel="Phase (degrees)",
     title="Bode Phase Plot", legend=false, grid=true)
```

# $f(x) = A e^{-\lambda x}$

(a) Prove that $A = \lambda$

## (a) Prove that $A = \lambda$

To prove that $A = \lambda$, we assume that $f(x)$ is a probability density function (PDF) of an exponential distribution. This means it must satisfy the **normalization condition**:

$\int_{0}^{\infty} f(x) \, dx = 1.$

### Step 1: Set Up the Integral
Given:

$f(x) = A e^{-\lambda x}, \quad x \geq 0.$

The total probability must be 1:

$\int_0^\infty A e^{-\lambda x} \, dx = 1.$

### Step 2: Evaluate the Integral
The integral of an exponential function is:

$\int e^{-\lambda x} \, dx = \frac{e^{-\lambda x}}{-\lambda}.$

Applying the limits:

$\int_0^\infty A e^{-\lambda x} \, dx = A \left[ \frac{e^{-\lambda x}}{-\lambda} \right]_{0}^{\infty}.$

At $x = \infty$, $e^{-\lambda x} \to 0$, and at $x = 0$, $e^{-\lambda \cdot 0} = 1$. Thus:

$A \left( 0 - \left(-\frac{1}{\lambda}\right) \right) = A \cdot \frac{1}{\lambda} = 1.$

### Step 3: Solve for $A$
$A = \lambda.$

### Conclusion:
Thus, the normalization condition requires that:

$A = \lambda.$

## (b) Find the Charasteristic Function 
        
The **characteristic function** of a real-valued random variable $X$ is defined as the expected value of $e^{itX}$, where $t$ is a real number:

$\varphi_X(t) = \mathbb{E}[e^{itX}] = \int_{-\infty}^{\infty} e^{itx} f_X(x) \, dx.$

### Explanation:
- $\varphi_X(t)$ uniquely characterizes the probability distribution of $X$.
- It is the **Fourier transform** of the probability density function (PDF) $f_X(x)$ (if $X$ is continuous).
- If $X$ is a discrete random variable, the summation replaces the integral:

$\varphi_X(t) = \sum_{x} e^{itx} P(X = x).$

### Properties of Characteristic Functions:
1. **$\varphi_X(0) = 1$** (since expectations preserve probability mass).
2. **Always exists** (unlike the moment-generating function, which may not be finite for all values).
3. **$|\varphi_X(t)| \leq 1$** for all $t$.
4. **Uniquely determines the distribution** of $X$.
5. **For independent random variables $X$ and $Y$**, $\varphi_{X+Y}(t) = \varphi_X(t) \varphi_Y(t)$.

It plays a crucial role in probability theory, especially in **limit theorems** like the **Central Limit Theorem (CLT)**.

The characteristic function of a random variable $X$ with probability density function (PDF) $f(x)$ is defined as:

$\varphi_X(t) = \mathbb{E}[e^{itX}] = \int_{-\infty}^{\infty} e^{itx} f(x) \, dx.$

For an **exponential distribution** with PDF:

$f(x) = \lambda e^{-\lambda x}, \quad x \geq 0,$

we compute:

$\varphi_X(t) = \int_0^\infty e^{itx} \lambda e^{-\lambda x} \, dx.$

### Step 1: Factor the Exponential
$\varphi_X(t) = \lambda \int_0^\infty e^{(it - \lambda)x} \, dx.$

### Step 2: Compute the Integral
The integral:

$\int_0^\infty e^{(it - \lambda)x} \, dx$

is a standard form:

$\int_0^\infty e^{-\alpha x} \, dx = \frac{1}{\alpha}, \quad \text{for } \text{Re}(\alpha) > 0.$

Here, $\alpha = \lambda - it$, so:

$\int_0^\infty e^{(it - \lambda)x} \, dx = \frac{1}{\lambda - it}.$

Multiplying by $\lambda$, we get:

$\varphi_X(t) = \frac{\lambda}{\lambda - it}.$

### Conclusion:
Thus, the characteristic function of the exponential distribution is:

$\varphi_X(t) = \frac{\lambda}{\lambda - it}, \quad \text{for } t \in \mathbb{R}.$
