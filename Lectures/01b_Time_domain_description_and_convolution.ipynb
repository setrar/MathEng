### Common Laplace Transforms

| Time Domain $ f(t) $         | Laplace Transform $ F(s) $         |
|---------------------------------|---------------------------------------|
| $ 1 $                         | $ \frac{1}{s} $                    |
| $ t $                         | $ \frac{1}{s^2} $                  |
| $ t^n $                       | $ \frac{n!}{s^{n+1}} $             |
| $ e^{\alpha t} $              | $ \frac{1}{s + \alpha} $                |
| $ \sin(\omega t) $          | $ \frac{\omega}{s^2 + \omega^2} $ |
| $ \cos(\omega t) $          | $ \frac{s}{s^2 + \omega^2} $      |
| $ e^{at} \sin(\omega t) $   | $ \frac{\omega}{(s - a)^2 + \omega^2} $ |
| $ e^{at} \cos(\omega t) $   | $ \frac{s - a}{(s - a)^2 + \omega^2} $ |
| $ u(t) $ (unit step function) | $ \frac{1}{s} $                    |
| $ \delta(t) $ (impulse function) | $ 1 $                          |

### Properties of Laplace Transforms

1. **Linearity**:
   $
   \mathcal{L}\{af(t) + bg(t)\} = aF(s) + bG(s)
   $

2. **Time Shifting**:
   $
   \mathcal{L}\{e^{at}f(t)\} = F(s - a)
   $

3. **Frequency Shifting**:
   $
   \mathcal{L}\{f(t)e^{at}\} = F(s - a)
   $

4. **Differentiation**:
   $
   \mathcal{L}\left\{ \frac{d}{dt} f(t) \right\} = sF(s) - f(0)
   $

5. **Integration**:
   $
   \mathcal{L}\left\{ \int_0^t f(\tau) d\tau \right\} = \frac{F(s)}{s}
   $

6. **Final Value Theorem**:
   $
   \lim_{t \to \infty} f(t) = \lim_{s \to 0} sF(s)
   $

7. **Initial Value Theorem**:
   $
   f(0) = \lim_{s \to \infty} sF(s)
   $

### Inverse Laplace Transforms

The inverse Laplace transform can be represented as:

$
\mathcal{L}^{-1}\{F(s)\} = \frac{1}{2\pi j} \int_{c - j\infty}^{c + j\infty} e^{st} F(s) ds
$

where $ c $ is a constant chosen such that all singularities of $ F(s) $ are to the left of the line $ \text{Re}(s) = c $.

This table and properties provide a foundation for using Laplace transforms in solving differential equations and analyzing linear time-invariant systems.
