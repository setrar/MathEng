# Reflected Paths

**Reflected sound paths** are a key factor in producing **reverb** when a string is plucked on a musical instrument. Reverb (short for reverberation) occurs when sound waves reflect off surfaces and return to the listener, combining with the original sound. This effect can create a sense of space and depth, commonly heard in large rooms, concert halls, or when amplified through equipment with a reverb effect.

### How It Works:

1. **Initial Sound**: When a string is plucked, it produces a sound wave that travels outward in all directions. This direct sound reaches the listener first.

2. **Reflected Paths**: The sound waves also reflect off nearby surfaces—walls, floors, ceilings, or the body of the instrument itself (like a guitar or violin). Each reflection takes a different path and arrives at the listener after some delay, depending on the distance traveled.

3. **Combining Direct and Reflected Waves**: The combination of the direct sound and these multiple reflections arriving at slightly different times creates what we hear as reverb. The timing, amplitude, and frequency characteristics of the reflections determine the nature of the reverb:
   - **Early Reflections**: These are the first few reflections that arrive after the direct sound and give clues about the size of the space.
   - **Late Reflections**: These are more scattered and build up into a wash of sound that decays over time.

4. **Natural vs. Artificial Reverb**: 
   - **Natural Reverb** occurs in physical spaces like rooms, halls, or outdoors when sound reflects off various surfaces.
   - **Artificial Reverb** can be created using digital effects processors or amplifiers, mimicking the reflections and reverberations you’d find in physical spaces.

### Physical Example with a String Instrument:
In the case of an acoustic guitar, for example:
- **Initial Vibration**: The string vibrates, producing sound waves that resonate through the body of the guitar.
- **Reflections in the Guitar Body**: The sound waves reflect within the hollow body of the guitar, reinforcing certain frequencies and decaying over time. This adds warmth and richness to the sound.
- **Room Reflections**: When played in a room, sound waves bounce off the walls, ceiling, and other surfaces, creating reverb that adds to the richness of the sound.

### Digital Reverb:
When using an electric guitar with an amplifier or digital effect processor, artificial reverb effects mimic these reflections and reverberations, adding a sense of space to the sound, even if you're playing in a "dry" (non-reverberant) room.

### Reverb Types:
- **Room Reverb**: Simulates the sound of a small room with many reflections.
- **Hall Reverb**: Simulates larger spaces like concert halls, with a longer decay.
- **Plate and Spring Reverb**: Created artificially with mechanical devices, often used in guitar amplifiers.

### Conclusion:
Reverb is directly related to **reflected sound paths** and plays a crucial role in shaping the tonal quality of plucked string instruments. It gives a sense of space and ambiance, whether occurring naturally in a physical environment or being artificially generated through sound processing.

```julia
using WAV, Plots

# Parameters for the sound
fs = 44100          # Sampling frequency (44.1 kHz)
f0 = 440            # Fundamental frequency (A4 note, 440 Hz)
duration = 2.0      # Duration of the sound in seconds
t = 0:1/fs:duration # Time vector

# Generate the original plucked string sound (simple sine wave)
pluck_sound = sin.(2π * f0 * t)

# Function to apply reverb
function add_reverb(signal, fs, decay_factor=0.5, delay_ms=50)
    delay_samples = round(Int, delay_ms / 1000 * fs)
    output = copy(signal)
    
    for i in delay_samples+1:length(signal)
        output[i] += decay_factor * signal[i - delay_samples]
    end
    
    return output
end

# Apply reverb to the plucked sound
reverb_sound = add_reverb(pluck_sound, fs, 0.5, 50)

# Save the reverberated sound as a WAV file
wavwrite(reverb_sound, fs, "data/reverberated_string_440hz.wav")

println("Reverberated sound saved as 'data/reverberated_string_440hz.wav'")

# Plotting the original and reverberated sound waves for visualization
plot(t[1:2000], pluck_sound[1:2000], label="Original", xlabel="Time (s)", ylabel="Amplitude", title="Plucked String (440 Hz)")
plot!(t[1:2000], reverb_sound[1:2000], label="With Reverb")

```

```julia
using WAV, Plots

# Parameters for the sound
fs = 44100          # Sampling frequency (44.1 kHz)
f0 = 440            # Fundamental frequency (A4 note, 440 Hz)
duration = 2.0      # Duration of the sound in seconds
t = 0:1/fs:duration # Time vector

# Generate the original plucked string sound (simple sine wave)
pluck_sound = sin.(2π * f0 * t)

# Function to apply multiple layers of reverb
function add_reverb(signal, fs, decay_factors=[0.6, 0.3], delays_ms=[50, 120])
    output = copy(signal)
    
    # Add reverb for each decay/delay combination
    for (decay_factor, delay_ms) in zip(decay_factors, delays_ms)
        delay_samples = round(Int, delay_ms / 1000 * fs)
        for i in delay_samples+1:length(signal)
            output[i] += decay_factor * signal[i - delay_samples]
        end
    end
    
    return output
end

# Apply more complex reverb to the plucked sound
reverb_sound = add_reverb(pluck_sound, fs, [0.6, 0.4, 0.2], [50, 120, 200])

# Save the reverberated sound as a WAV file
wavwrite(reverb_sound, fs, "data/reverberated_string_440hz_more_reverb.wav")

println("Reverberated sound saved as 'data/reverberated_string_440hz_more_reverb.wav'")

# Plotting the original and reverberated sound waves for visualization
plot(t[1:2000], pluck_sound[1:2000], label="Original", xlabel="Time (s)", ylabel="Amplitude", title="Plucked String (440 Hz)")
plot!(t[1:2000], reverb_sound[1:2000], label="With Reverb")

```
