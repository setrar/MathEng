## &#x274D;  Signal classification

- many ways in which signal may be classified, i.e. periodic, when $x(t) = x(t + T )$, where the smallest value of $T$ defines the period
- we can also classify signals as either energy or power signals

| |
|-|
|<pre style="border:1px solid yellow; padding:10px;"> finite energy = zero power <br> finite power = infinite energy </pre> |

- energy signals
    - non-zero and finite total dissipated energy, $E$
    - usually exist for a finite interval of time or have most of their energy concentrated in a finite interval of time
      $0 \leq E \leq \infty, E = \int_{- \infty}^{\infty} x^2(t) dt$


- power signals
    - on-zero and finite average delivered power, $P$, i.e. $0 < P < \infty $
    - an example is the unit step function $u(t)$ and a periodic signal of period $T$ such as $x(t)=sin(2 \pi t / T)$
$$
\frac{1}{t_2 - t_1} \int_{t_1}^{t_2} x^2(t) \it{d}t \quad P = \lim_{T -> \infty} \frac{1}{2T} \int_{-T}^{T} x^2(t) \it{d}t \quad P = \frac{1}{T} \int_{0}^{T} x^2(t) \it{d}t
$$

---

#### Example
##### Find the energy in the decaying exponential signal $x_1(t)=5 \exp(-2t)$ if $t \geq 0$ and $x_1(t) = 0$ if $t < 0$.

---

We want to find the energy of the signal $ x_1(t) = 5 \exp(-2t) $ for $ t \geq 0 $ and $ x_1(t) = 0 $ for $ t < 0 $, using the energy formula:

$
E = \int_{-\infty}^{\infty} |x_1(t)|^2 \, dt
$

Since $ x_1(t) = 5 \exp(-2t) $ for $ t \geq 0 $, the integral becomes:

$
E = \int_{0}^{\infty} |5 \exp(-2t)|^2 \, dt
$

Squaring the amplitude gives:

$
|5 \exp(-2t)|^2 = (5 \exp(-2t))^2 = 25 \exp(-4t)
$

Now, the energy is:

$
E = \int_{0}^{\infty} 25 \exp(-4t) \, dt
$

### Solving the Integral

1. Factor out the constant $ 25 $:
   $
   E = 25 \int_{0}^{\infty} \exp(-4t) \, dt
   $

2. The integral of $ \exp(at) $ is $ \frac{1}{a} \exp(at) $. So:
   $
   \int \exp(-4t) \, dt = \frac{1}{-4} \exp(-4t)
   $

3. Apply the limits of integration from $ 0 $ to $ \infty $:
   $
   E = 25 \left[ \frac{1}{-4} \exp(-4t) \right]_{0}^{\infty}
   $

4. Evaluate at the limits:
   - As $ t \to \infty $, $ \exp(-4t) \to 0 $
   - At $ t = 0 $, $ \exp(-4(0)) = 1 $
   
   Therefore:
   $
   E = 25 \left( 0 - \frac{1}{-4} \right)
   $
   $
   E = 25 \times \frac{1}{4}
   $
   $
   E = \frac{25}{4} = 6.25
   $

### Final Answer:
The total energy of the signal $ x_1(t) = 5 \exp(-2t) $ is:

$
E = 6.25
$ units of energy.

#### Note
The expressions $ 5 e^{-2t} $ and $ 5 \exp(-2t) $ are equivalent. $ e^{-2t} $ uses traditional mathematical notation, while $ \exp(-2t) $ is a programming-friendly form. Both represent the same exponential decay function.

## &#x274D; Fourier Series

### &#x27B0; Trigonometric Fourier series

- we can represent any finite power periodic signal $x(t)$ with a period $T$as a sum of sine and cosine waves:

$$
x(t) = \frac{A_0}{2} + \sum_{n=1}^\infty A_n \cos(n \omega_0 t) + B_n \sin(n \omega_0 t)
$$

$ \omega_0 $ is:

- **$ \omega_0 $**, the **fundamental angular frequency** is measured in **radians per second** $(\text{rad/s})$.
  
  Since $ \omega_0 = 2\pi f_0 $, the factor of $ 2\pi $ converts the frequency from cycles per second (Hertz) to radians per second. One complete cycle corresponds to $ 2\pi $ radians.

- $ f_0 = \frac{1}{T} $ the **fundamental frequency**. is measured in **Hertz (Hz)**, which is equivalent to **cycles per second** $(\text{s}^{-1})$.

Therefore:
- $ f_0 $ represents how many complete cycles occur per second.
- $ \omega_0 $ represents how many radians of phase change occur per second.

- This frequency $ f_0 $ corresponds to the **lowest harmonic** in the sum, with higher harmonics being integer multiples of $ f_0 $.
- harmonics are generally found at 2/T Hz, 3/T Hz ... according to Fourier coefficients:

$$
A_n = \frac{2}{T} \int_{\frac{-T}{2}}^{\frac{T}{2}} x(t) cos(n \omega_0 t) \; \mathrm{d}t \quad n = 0, 1, 2, ...
$$

$$
B_n = \frac{2}{T} \int_{\frac{-T}{2}}^{\frac{T}{2}} x(t) sin(n \omega_0 t) \; \mathrm{d}t \quad n = 1, 2, 3, ...
$$

### &#x27B0; Complex phasors

Sine and cosine waves can be described using **complex phasors** because complex numbers provide a convenient way to represent oscillatory signals with both amplitude and phase. Phasors allow us to encapsulate the behavior of sinusoidal functions in a simple, compact form, which is especially useful in fields like signal processing, control systems, and electrical engineering.

Here’s why and how this works:

### 1. Euler's Formula

A key reason sine and cosine waves can be represented as complex phasors is **Euler's formula**, which states:

$$
e^{j \theta} = \cos(\theta) + j \sin(\theta),
$$

where $ j $ (or $ i $ in some conventions) is the imaginary unit, with $ j^2 = -1 $. Using this formula, any sinusoidal function can be expressed in terms of complex exponentials.

### 2. Representing Sine and Cosine Waves with Complex Exponentials

Consider a cosine wave with angular frequency $ \omega $ and phase $ \phi $:

$
\cos(\omega t + \phi).
$

Using Euler's formula, we can rewrite this cosine wave as:

$
\cos(\omega t + \phi) = \operatorname{Re} \left\{ e^{j(\omega t + \phi)} \right\} = \operatorname{Re} \left\{ e^{j \phi} \cdot e^{j \omega t} \right\}.
$

In this form:
- $ e^{j \phi} $ is a **phasor** representing the initial phase of the wave.
- $ e^{j \omega t} $ represents the time evolution of the wave.

A sine wave can similarly be expressed as the imaginary part of a complex exponential:

$
\sin(\omega t + \phi) = \operatorname{Im} \left\{ e^{j(\omega t + \phi)} \right\}.
$

### 3. Advantages of Using Phasors

Representing sinusoidal signals with phasors $ e^{j \omega t} $ has several benefits:

- **Simplifies Calculations**: Using phasors, we can add, subtract, or multiply sinusoidal signals by simply working with complex numbers, avoiding the need to manipulate trigonometric identities.
  
- **Captures Amplitude and Phase**: A phasor represents both the **amplitude** and **phase** of a sinusoid in a single complex number. For example, a phasor $ Ae^{j \phi} $ has an amplitude $ A $ and phase $ \phi $, corresponding to the real sinusoid $ A \cos(\omega t + \phi) $.

- **Time-Derivative Simplicity**: In differential equations, the time derivative of a sinusoidal signal becomes multiplication by $ j \omega $ when using phasors, which simplifies analysis, particularly in electrical circuits and systems with oscillatory behavior.

### 4. Example of a Phasor Representation

Suppose we have a cosine wave:

$
x(t) = A \cos(\omega t + \phi).
$

We can represent it as a complex phasor:

$
X = A e^{j \phi},
$

where $ X $ is a complex constant (the phasor), and the time-dependent wave can be reconstructed as:

$
x(t) = \operatorname{Re} \left\{ X e^{j \omega t} \right\}.
$

### Summary

Using complex phasors, sine and cosine waves can be represented in a compact and efficient form, leveraging the properties of complex exponentials to handle amplitude, phase, and frequency with ease. This approach is particularly powerful in engineering and physics, where phasors streamline the analysis of oscillatory systems.

### &#x27B0; Complex Fourier series

- substituting the last two equations on the last slide into those for the
trigonometric Fourier series gives us the complex Fourier series:

$$
x(t) = \sum\limits_{n = -\infty}^{+\infty} X_n e^{j n \omega_0 t}
$$

    which we can interpret as a bank of phasor generators with increasing frequencies and amplitudes $X_n$

$$
X_n = \frac{1}{T} \int_{T} x(t) e^{-j n \omega_0 t } \, dt
$$

- one difference between the trigonometric and the complex Fourier series – the trigonometric Fourier series has three equations whereas the complex Fourier series has only two but $X_ns$ in the complex form are often complex

Derive an expression for the complex Fourier coefficient, Xn, associated with

```julia
# Import the necessary library
using Plots

# Define parameters
A = 1.0        # Amplitude of the pulse
T = 3.3        # Period of the signal
τ = 1.0        # Pulse width
num_periods = 6  # Number of periods to plot

# Define the signal x(t)
function x(t, A, T, τ)
    # Compute the position within the period
    t_mod = t % T
    # Check if t_mod is within the pulse width τ
    if abs(t_mod - T/2) <= τ/2
        return A
    else
        return 0.0
    end
end

# Generate data for plotting
t_values = range(-num_periods*T/2, stop=num_periods*T/2, length=1000)  # Time values
x_values = [x(t, A, T, τ) for t in t_values]                           # Signal values

# Plot x(t)
plot(t_values, x_values
    , linewidth=2, legend=:false, size = (600,300)
    , xlabel="t", ylabel="x(t)"
    , xlim = (0, 10), xticks=:false
    , title="Periodic Pulse Train Signal x(t)")
vline!([1.65, 5, 8.25], linestyle=:dash, color=:gray, label=:none)  # Dashed gray lines
```

To derive the expression for the complex Fourier coefficient $ X_n $ for the periodic signal $ x(t) $ shown in the image, we can follow these steps:

### Step 1: Define the Fourier Coefficient
The complex Fourier coefficient $ X_n $ for a periodic signal $ x(t) $ with period $ T $ is given by:
$
X_n = \frac{1}{T} \int_{T} x(t) e^{-j 2 \pi n \frac{t}{T}} \, dt
$
where $ n $ is an integer representing the harmonic number.

### Step 2: Identify the Signal Characteristics
From the image, the signal $ x(t) $ appears to be a periodic pulse train with a period $ T $. Each pulse has an amplitude $ A $ and a width $ \tau $, centered around multiples of $ T $. The pulse duration is therefore $ \tau $, and it is symmetric around each interval center.

### Step 3: Set up the Integral over One Period
Since the signal is periodic, we can compute $ X_n $ by integrating over one period. We'll assume that the pulse is centered at $ t = 0 $ and extends from $ -\frac{\tau}{2} $ to $ \frac{\tau}{2} $.

Thus, the expression for $ X_n $ becomes:
$
X_n = \frac{1}{T} \int_{-\frac{\tau}{2}}^{\frac{\tau}{2}} A \, e^{-j 2 \pi n \frac{t}{T}} \, dt
$

### Step 4: Evaluate the Integral
Now, we evaluate the integral:
$
X_n = \frac{A}{T} \int_{-\frac{\tau}{2}}^{\frac{\tau}{2}} e^{-j 2 \pi n \frac{t}{T}} \, dt
$

Since $ A $ and $ T $ are constants, they can be factored out of the integral:
$
X_n = \frac{A}{T} \left[ \frac{T}{-j 2 \pi n} e^{-j 2 \pi n \frac{t}{T}} \right]_{-\frac{\tau}{2}}^{\frac{\tau}{2}}
$

Evaluating this expression at the bounds $ t = \pm \frac{\tau}{2} $, we get:
$
X_n = \frac{A}{T} \cdot \frac{T}{-j 2 \pi n} \left( e^{-j 2 \pi n \frac{\tau}{2T}} - e^{j 2 \pi n \frac{\tau}{2T}} \right)
$

Using Euler's formula, we can simplify this to:
$
X_n = \frac{A}{T} \cdot \frac{T}{-j 2 \pi n} \cdot \left( -2j \sin\left( \pi n \frac{\tau}{T} \right) \right)
$

Simplifying further, we get:
$
X_n = A \frac{\tau}{T} \frac{\sin\left( \pi n \frac{\tau}{T} \right)}{\pi n \frac{\tau}{T}}
$

### Step 5: Final Expression
This gives the complex Fourier coefficient for the signal:
$
X_n = A \frac{\tau}{T} \text{sinc} \left( n \frac{\tau}{T} \right)
$
where $ \text{sinc}(x) = \frac{\sin(\pi x)}{\pi x} $.

This expression provides the Fourier coefficients $ X_n $ for the given periodic pulse train signal $ x(t) $.

## &#x274D; Transform analysis of linear systems

- the transforms all represent signals as weighted sums (integrals) of
exponential orthogonal basis functions

    – e.g. for the complex Fourier series we have weights $X_n$ and
    mutually orthogonal basis functions $e^{(j n \omega_0 t)}$

- this representation is fundamental to the analysis of linear systems
and to evaluating the response of such systems to a wide range of
inputs

    – superposition: a system with a number of inputs has an output
    equal to the sum of the output of each input

### &#x27B0; Linear ordinary differential equations

The equation you've provided resembles a linear differential equation that describes the relationship between the output $ y(t) $ and the input $ x(t) $ of a dynamic system. Here's a breakdown of its components:

1. **General Form**:
   $
   a_0 y + a_1 \frac{dy}{dt} + \cdots + a_n \frac{d^n y}{dt^n} = b_0 x + b_1 \frac{dx}{dt} + \cdots + b_m \frac{d^m x}{dt^m}
   $
   This is a linear ordinary differential equation (ODE) of order $ n $ in terms of the output $ y(t) $ and order $ m $ in terms of the input $ x(t) $.

2. **Terms Explanation**:
   - $ a_0, a_1, \ldots, a_n $: Coefficients of the output $ y(t) $ and its derivatives.
   - $ b_0, b_1, \ldots, b_m $: Coefficients of the input $ x(t) $ and its derivatives.
   - $ \frac{dy}{dt}, \frac{d^n y}{dt^n} $: Derivatives of the output with respect to time.
   - $ \frac{dx}{dt}, \frac{d^m x}{dt^m} $: Derivatives of the input with respect to time.

3. **Interpretation**:
   - This equation represents a linear time-invariant (LTI) system if the coefficients $ a_i $ and $ b_j $ are constants.
   - It describes how the output $ y(t) $ responds to the input $ x(t) $ based on its own past behavior (through its derivatives).

4. **Applications**:
   - Such equations are commonly used in control systems, electrical circuits, mechanical systems, and many other fields of engineering and physics to model the dynamics of systems.

5. **Solution Methods**:
   - The solution can often be found using techniques such as the Laplace transform or numerical methods, depending on the nature of the coefficients and initial conditions.

In summary, this equation represents a linear relationship between input and output in a dynamic system, capturing how changes in the input affect the output over time through derivatives.
