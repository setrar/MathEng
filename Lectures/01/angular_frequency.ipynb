**Angular frequency** is a measure of how quickly an object rotates or oscillates in terms of the angle covered per unit of time. It is commonly used to describe the oscillation rate of sinusoidal waves in fields like physics, engineering, and signal processing.

### Definition of Angular Frequency

Angular frequency, usually denoted by $ \omega $, is defined as:

$
\omega = 2 \pi f,
$

where:
- $ f $ is the **frequency** of the oscillation in cycles per second (Hertz, or Hz),
- $ \omega $ is the angular frequency in **radians per second** (rad/s).

### Explanation and Interpretation

- **Units**: Angular frequency is measured in radians per second. Since there are $ 2 \pi $ radians in one complete cycle (or 360 degrees), multiplying by $ 2 \pi $ converts frequency from cycles per second (Hertz) to radians per second.
- **Meaning**: Angular frequency tells us how much "angle" an oscillating object (like a point on a rotating wheel or an oscillating wave) covers per second. In the case of a wave, it indicates the rate of oscillation in terms of radians.

For example:
- If $ f = 1 $ Hz, then $ \omega = 2 \pi $ rad/s, meaning the wave completes $ 2 \pi $ radians (one full cycle) per second.
- If $ f = 60 $ Hz, then $ \omega = 2 \pi \cdot 60 = 120 \pi $ rad/s, which is the angular frequency of a typical 60 Hz AC power signal.

### Angular Frequency in Sinusoidal Waves

In sinusoidal waves, angular frequency is critical for describing the waveâ€™s oscillatory behavior. A sinusoidal function with angular frequency $ \omega $ is often written as:

$
x(t) = A \cos(\omega t + \phi),
$

where:
- $ A $ is the amplitude,
- $ \omega $ determines how quickly the oscillations happen,
- $ \phi $ is the phase offset.

Here, $ \omega t $ gives the phase of the wave at time $ t $. As $ t $ increases, the phase progresses, causing the wave to oscillate.

### Relation to Period

The **period** $ T $ of a wave (the time it takes to complete one cycle) is inversely related to the frequency $ f $:

$
f = \frac{1}{T}.
$

Since $ \omega = 2 \pi f $, we can also express angular frequency in terms of the period:

$
\omega = \frac{2 \pi}{T}.
$

### Applications of Angular Frequency

Angular frequency is used in many applications:
- **Oscillating Circuits**: In AC circuits, the voltage and current oscillate with a certain angular frequency, which is determined by the circuit's components.
- **Rotational Motion**: In rotational systems, angular frequency describes how fast an object rotates.
- **Signal Processing**: Angular frequency is used to analyze signals in the frequency domain, especially when working with Fourier series or Fourier transforms.

### Summary

Angular frequency $ \omega $ is a measure of how fast something oscillates or rotates, expressed in radians per second. It is a fundamental concept in wave mechanics, oscillating systems, and signal processing, providing a natural way to describe the rate of cyclical processes in terms of angle per unit time.

The maximum radian frequency $ \omega_{\text{max}} $ observed due to Doppler shift can be calculated based on the maximum Doppler-shifted frequency.

From the Doppler shift formula in the previous notes:

$
f_{\text{effective}} = f \left( 1 - \frac{v}{c} \right)
$

The Doppler shift $ D $ in frequency is:

$
D = -\frac{f v}{c}
$

Thus, the Doppler-shifted frequency is:

$
f_{\text{observed}} = f + D = f \left( 1 - \frac{v}{c} \right)
$

To find the maximum angular frequency $ \omega_{\text{max}} $ in radians per second, we convert the maximum observed frequency to radians:

$
\omega_{\text{max}} = 2 \pi f_{\text{observed}} = 2 \pi f \left( 1 - \frac{v}{c} \right)
$

The maximum angular frequency $ \omega_{\text{max}} $ occurs when the relative velocity $ v $ is at its maximum possible value. In practical scenarios, the maximum $ v $ would be limited by the speed of light $ c $, but for objects moving at significantly lower speeds than $ c $, the Doppler effect will still produce a measurable shift in $ \omega $ based on the velocity.

### Summary Formula for Maximum Angular Frequency

$
\omega_{\text{max}} = 2 \pi f \left( 1 - \frac{v_{\text{max}}}{c} \right)
$

where $ v_{\text{max}} $ is the maximum relative velocity between the transmitter and receiver. This formula provides the maximum angular frequency observed, taking into account the Doppler shift.

# References

The notation $ E \propto \frac{1}{r} $ means that the **electric field strength $ E $** is **inversely proportional to the distance $ r $** from the source. This implies:

1. **Field Strength Decreases with Distance**: As the distance $ r $ from the source increases, the electric field $ E $ decreases. Specifically, if you double the distance $ r $, the electric field strength will be reduced by half.

2. **Spherical Wave Propagation**: This $ \frac{1}{r} $ relationship is characteristic of **spherical wave propagation** in the **far field**. When a wave radiates outward from a point source, it spreads over a spherical surface, and the intensity of the field decreases as the surface area of the sphere (which grows as $ 4 \pi r^2 $) increases.

3. **Energy Spreading**: Since the total energy radiated by the source is spread over a larger and larger area as $ r $ increases, the energy (and hence the field strength) per unit area decreases proportionally to $ \frac{1}{r^2} $. However, the electric field $ E $, which is related to the square root of power per unit area, decreases as $ \frac{1}{r} $.

In summary, $ E \propto \frac{1}{r} $ indicates that the electric field strength weakens as you move farther from the source, due to the spreading of the wave over a larger area in the far field.
