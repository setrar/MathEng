A **complex phasor** is a way to represent a sinusoidal signal in terms of its **amplitude** and **phase** using a complex number. Phasors simplify the analysis of oscillating signals, particularly in electrical engineering, physics, and signal processing, by converting trigonometric functions (sines and cosines) into a more manageable complex exponential form.

### Definition of a Complex Phasor

Mathematically, a complex phasor is often represented as:

$
X = A e^{j \phi},
$

where:
- $ A $ is the **magnitude** (or amplitude) of the sinusoidal signal,
- $ \phi $ is the **phase angle** of the signal,
- $ j $ (or $ i $ in some fields) is the imaginary unit, with $ j^2 = -1 $.

### Relationship to Sinusoidal Signals

For a sinusoidal signal $ x(t) = A \cos(\omega t + \phi) $, we can represent it using Euler's formula:

$
x(t) = A \cos(\omega t + \phi) = \operatorname{Re} \left\{ A e^{j(\omega t + \phi)} \right\}.
$

In phasor form:
- $ A e^{j \phi} $ is the **complex phasor** that represents the amplitude $ A $ and phase $ \phi $ of the sinusoid.
- $ e^{j \omega t} $ describes the **time evolution** of the oscillation, with frequency $ \omega $.

### Benefits of Using Phasors

Using complex phasors to represent sinusoidal signals has several advantages:

1. **Simplifies Analysis**: In systems where multiple sinusoidal signals interact (such as in AC circuit analysis), phasors allow us to add, subtract, and multiply signals as complex numbers rather than dealing with trigonometric identities.
  
2. **Captures Amplitude and Phase**: A phasor represents both the **magnitude** and **phase** in a single complex number, making it easier to track both properties without separately considering cosine and sine components.

3. **Simplifies Calculus Operations**: When taking derivatives or integrals of sinusoidal signals, using phasors allows these operations to be handled as simple multiplications or divisions by $ j \omega $, which is especially useful in differential equations and control systems.

### Example of Phasor Representation

Suppose we have a cosine signal:

$
x(t) = 5 \cos(2 \pi \cdot 60 \, t + \frac{\pi}{4}).
$

We can represent this signal as a complex phasor:

1. The amplitude $ A = 5 $.
2. The phase $ \phi = \frac{\pi}{4} $.
3. The frequency $ \omega = 2 \pi \cdot 60 $ rad/s.

So, the phasor representation is:

$
X = 5 e^{j \frac{\pi}{4}}.
$

The time-dependent sinusoid can then be written as:

$
x(t) = \operatorname{Re} \left\{ 5 e^{j \frac{\pi}{4}} e^{j 2 \pi \cdot 60 \, t} \right\}.
$

### Summary

A **complex phasor** is a compact, complex-number representation of a sinusoidal signal, capturing both its amplitude and phase. By transforming sinusoids into phasors, we can greatly simplify mathematical operations, particularly in fields involving oscillatory phenomena like AC circuits, vibrations, and signal processing.

```julia
using Plots

# Define the phasor parameters
A = 5.0                    # Amplitude
f = 60.0                   # Frequency in Hz
ω₀ = 2 * π * f              # Angular frequency in rad/s
ϕ = π / 4                  # Phase in radians

# Define the phasor as a complex number
X = A * ℯ^(j * ϕ)        # Complex phasor representation

# Define the time vector over one period of the signal
tᵥ = 0:0.0001:(1/f)         # Time vector from 0 to one period of the signal

# Calculate x(t) using the phasor representation
xₜ = [𝓡𝓮(X * ℯ^(j * ω₀ * t)) for t in tᵥ]

# Plot the result
plot(tᵥ, xₜ
    , label="x(t) = 5 cos(2π * 60 t + π/4)"
    , xlabel="Time (s)", ylabel="x(t)"
    , title="Phasor Representation of a Sinusoidal Signal"
)
```
