Example: 
Two random variables $X$ and $Y$ have the joint pdf

$$
f_{XY}(x,y) =
  \begin{cases}
    A e^-(2x+y)       & \quad , x, y \geq 0 \\
    0  & \quad \text{otherwise }
  \end{cases}
$$

where $A$ is a constant. Determine $A$ and find the two maginal pdf's.

---
To solve for $ A $ and find the marginal PDFs of $ X $ and $ Y $, we proceed as follows:

### Step 1: Determine $ A $
Since $ F_{XY}(x, y) $ is a joint probability density function, it must integrate to 1 over the entire support of $ x $ and $ y $. Thus, we have:
$
\int_0^{\infty} \int_0^{\infty} A e^{-(2x + y)} \, dx \, dy = 1
$

1. **Integrate with respect to $ x $:**
   $
   \int_0^{\infty} A e^{-(2x + y)} \, dx = A \int_0^{\infty} e^{-2x} e^{-y} \, dx = A e^{-y} \int_0^{\infty} e^{-2x} \, dx
   $
   The inner integral with respect to $ x $ is:
   $
   \int_0^{\infty} e^{-2x} \, dx = \left[ -\frac{1}{2} e^{-2x} \right]_0^{\infty} = \frac{1}{2}
   $
   Therefore,
   $
   \int_0^{\infty} A e^{-(2x + y)} \, dx = A e^{-y} \cdot \frac{1}{2} = \frac{A}{2} e^{-y}
   $

2. **Integrate with respect to $ y $:**
   Now we integrate with respect to $ y $:
   $
   \int_0^{\infty} \frac{A}{2} e^{-y} \, dy = \frac{A}{2} \int_0^{\infty} e^{-y} \, dy
   $
   The integral of $ e^{-y} $ over $ y $ from 0 to $ \infty $ is:
   $
   \int_0^{\infty} e^{-y} \, dy = \left[ -e^{-y} \right]_0^{\infty} = 1
   $
   So,
   $
   \int_0^{\infty} \int_0^{\infty} A e^{-(2x + y)} \, dx \, dy = \frac{A}{2} \cdot 1 = \frac{A}{2}
   $
   Setting this equal to 1 to satisfy the requirement of a probability density, we get:
   $
   \frac{A}{2} = 1 \Rightarrow \boxed{ A = 2 }
   $

Thus, the joint PDF is:
$
F_{XY}(x, y) = 
\begin{cases} 
2 e^{-(2x + y)} & x, y \geq 0 \\ 
0 & \text{otherwise}
\end{cases}
$

### Step 2: Find the Marginal PDFs

The marginal PDFs $ fₓ(x) $ and $ fᵧ(y) $ can be found by integrating the joint PDF over the other variable.

#### Marginal PDF of $ X $:
To find $ fₓ(x) $, integrate $ F_{XY}(x, y) $ over $ y $:
$
fₓ(x) = \int_0^{\infty} 2 e^{-(2x + y)} \, dy
$
1. Separate terms involving $ y $:
   $
   fₓ(x) = 2 e^{-2x} \int_0^{\infty} e^{-y} \, dy
   $
2. Evaluate the integral:
   $
   \int_0^{\infty} e^{-y} \, dy = 1
   $
   Thus,
   $
   fₓ(x) = 2 e^{-2x} \cdot 1 = 2 e^{-2x}, \quad x \geq 0
   $
Therefore,
$
fₓ(x) = 
\begin{cases}
2 e^{-2x} & x \geq 0 \\
0 & \text{otherwise}
\end{cases}
$

#### Marginal PDF of $ Y $:
To find $ fᵧ(y) $, integrate $ F_{XY}(x, y) $ over $ x $:
$
fᵧ(y) = \int_0^{\infty} 2 e^{-(2x + y)} \, dx
$
1. Separate terms involving $ x $:
   $
   fᵧ(y) = 2 e^{-y} \int_0^{\infty} e^{-2x} \, dx
   $
2. Evaluate the integral:
   $
   \int_0^{\infty} e^{-2x} \, dx = \frac{1}{2}
   $
   Therefore,
   $
   fᵧ(y) = 2 e^{-y} \cdot \frac{1}{2} = e^{-y}, \quad y \geq 0
   $
So,
$
fᵧ(y) = 
\begin{cases}
e^{-y} & y \geq 0 \\
0 & \text{otherwise}
\end{cases}
$

### Final Answer

The constant $ A $ is $ 2 $, and the marginal PDFs are:
$
fₓ(x) = 
\begin{cases}
2 e^{-2x} & x \geq 0 \\
0 & \text{otherwise}
\end{cases}
$
and
$
fᵧ(y) = 
\begin{cases}
e^{-y} & y \geq 0 \\
0 & \text{otherwise}
\end{cases}
$

Example: 

To illustrate the processes of normalisation of joint pdf’s, finding
marginal from joint pdf’s and checking for statistical independent

$$
F_{XY}(x,y) =
  \begin{cases}
    \beta xy       & \quad , 0 \leq x \leq y, 0 \leq y \leq 4 \\
    0  & \quad \text{otherwise }
  \end{cases}
$$

Determine $\beta$ and find the two marginal pdf’s.

---
To illustrate the processes of normalization, finding marginal PDFs, and checking for statistical independence, let's go through the steps for the given joint PDF:

$
F_{XY}(x, y) = 
\begin{cases} 
\beta xy & 0 \leq x \leq y \leq 4 \\ 
0 & \text{otherwise}
\end{cases}
$

### Step 1: Determine $ \beta $

The joint PDF $ F_{XY}(x, y) $ must integrate to 1 over the entire support of $ x $ and $ y $. This means:
$
\int_0^4 \int_0^y \beta xy \, dx \, dy = 1
$

1. **Integrate with respect to $ x $:**
   $
   \int_0^y \beta xy \, dx = \beta \int_0^y x y \, dx = \beta y \int_0^y x \, dx
   $
   The integral of $ x $ with respect to $ x $ from 0 to $ y $ is:
   $
   \int_0^y x \, dx = \left[ \frac{x^2}{2} \right]_0^y = \frac{y^2}{2}
   $
   Thus,
   $
   \int_0^y \beta xy \, dx = \beta y \cdot \frac{y^2}{2} = \frac{\beta y^3}{2}
   $

2. **Integrate with respect to $ y $:**
   Now we integrate with respect to $ y $:
   $
   \int_0^4 \frac{\beta y^3}{2} \, dy = \frac{\beta}{2} \int_0^4 y^3 \, dy
   $
   The integral of $ y^3 $ with respect to $ y $ from 0 to 4 is:
   $
   \int_0^4 y^3 \, dy = \left[ \frac{y^4}{4} \right]_0^4 = \frac{4^4}{4} = 16
   $
   Therefore,
   $
   \int_0^4 \int_0^y \beta xy \, dx \, dy = \frac{\beta}{2} \cdot 16 = 8\beta
   $
   Setting this equal to 1 to satisfy the requirement of a probability density, we get:
   $
   8\beta = 1 \Rightarrow \boxed{ \beta = \frac{1}{8} }
   $

So, the joint PDF is:
$
F_{XY}(x, y) = 
\begin{cases} 
\frac{1}{8} xy & 0 \leq x \leq y \leq 4 \\ 
0 & \text{otherwise}
\end{cases}
$

### Step 2: Find the Marginal PDFs

The marginal PDFs $ fₓ(x) $ and $ fᵧ(y) $ are obtained by integrating the joint PDF over the other variable.

#### Marginal PDF of $ X $:
To find $ fₓ(x) $, integrate $ F_{XY}(x, y) $ over $ y $:
$
fₓ(x) = \int_x^4 \frac{1}{8} xy \, dy
$
1. Factor out terms not involving $ y $:
   $
   fₓ(x) = \frac{x}{8} \int_x^4 y \, dy
   $
2. Evaluate the integral:
   $
   \int_x^4 y \, dy = \left[ \frac{y^2}{2} \right]_x^4 = \frac{4^2}{2} - \frac{x^2}{2} = \frac{16 - x^2}{2} = 8 - \frac{x^2}{2}
   $
   Therefore,
   $
   fₓ(x) = \frac{x}{8} \left(8 - \frac{x^2}{2}\right) = x \left(1 - \frac{x^2}{16}\right)
   $
So,
$
fₓ(x) = 
\begin{cases}
x \left(1 - \frac{x^2}{16}\right) & 0 \leq x \leq 4 \\
0 & \text{otherwise}
\end{cases}
$

#### Marginal PDF of $ Y $:
To find $ fᵧ(y) $, integrate $ F_{XY}(x, y) $ over $ x $:
$
fᵧ(y) = \int_0^y \frac{1}{8} xy \, dx
$
1. Factor out terms not involving $ x $:
   $
   fᵧ(y) = \frac{y}{8} \int_0^y x \, dx
   $
2. Evaluate the integral:
   $
   \int_0^y x \, dx = \left[ \frac{x^2}{2} \right]_0^y = \frac{y^2}{2}
   $
   Therefore,
   $
   fᵧ(y) = \frac{y}{8} \cdot \frac{y^2}{2} = \frac{y^3}{16}
   $
So,
$
fᵧ(y) = 
\begin{cases}
\frac{y^3}{16} & 0 \leq y \leq 4 \\
0 & \text{otherwise}
\end{cases}
$

### Step 3: Check for Statistical Independence

To check if $ X $ and $ Y $ are independent, we see if the joint PDF $ F_{XY}(x, y) $ can be factored as the product of the marginal PDFs $ fₓ(x) $ and $ fᵧ(y) $.

The joint PDF is:
$
F_{XY}(x, y) = \frac{1}{8} xy
$

The product of the marginal PDFs is:
$
fₓ(x) \cdot fᵧ(y) = x \left(1 - \frac{x^2}{16}\right) \cdot \frac{y^3}{16}
$

Clearly, $ F_{XY}(x, y) \neq fₓ(x) \cdot fᵧ(y) $, so $ X $ and $ Y $ are **not independent**.

```julia
using Plots
using LaTeXStrings
```

```julia
# Define the marginal PDFs for X and Y
function fₓ(x)
    if 0 <= x <= 4
        return x * (1 - x^2 / 16)
    else
        return 0
    end
end

function fᵧ(y)
    if 0 <= y <= 4
        return y^3 / 16
    else
        return 0
    end
end
```

```julia
# Generate a range of values for plotting
x_vals = 0:0.01:4
y_vals = 0:0.01:4

# Compute the values of the marginal PDFs
fₓ_vals = [fₓ(x) for x in x_vals]
fᵧ_vals = [fᵧ(y) for y in y_vals]

# Plot fₓ(x)
p1 = plot(x_vals, fₓ_vals, label=L"f_X(x) = x(1 - x^2/16)", xlabel=L"x", ylabel=L"f_X(x)", 
     title="Marginal PDF of X", linewidth=2, legend=:topright)

# Plot fᵧ(y)
p2 = plot(y_vals, fᵧ_vals, label=L"f_Y(y) = y^3/16", xlabel=L"y", ylabel=L"f_Y(y)", 
     title="Marginal PDF of Y", linewidth=2, legend=:topright)

plot(p1,p2
    , layout = (2,1) 
    , size = (500,500)
)
```

Example:

Derive the pdf of the random variable Y defined by

$Y = - ( \frac{1}{\pi} ) \Theta + 1$

where the random variable $\Theta$ has a pdf given by

$$
f₀(\theta) =
  \begin{cases}
    1/(2 \pi)       & \quad , 0 \leq \theta \leq 2 \pi \\
    0  & \quad \text{otherwise }
  \end{cases}
$$


To derive the PDF of the random variable $ Y = -\frac{1}{\pi} \Theta + 1 $, where $\Theta$ has the given PDF $ f_\Theta(\theta) $, let's go through the following steps.

### Step 1: Understand the Transformation and Range of $ Y $

Since $ Y $ is defined in terms of $\Theta$ by:
$
Y = -\frac{1}{\pi} \Theta + 1
$
we can rearrange this equation to express $\Theta$ in terms of $ Y $:
$
\Theta = -\pi(Y - 1) = -\pi Y + \pi
$

Since $ \Theta $ has support on $ [0, 2\pi] $, let's determine the corresponding range for $ Y $:
1. When $ \theta = 0 $:
   $
   Y = -\frac{1}{\pi} \cdot 0 + 1 = 1
   $
2. When $ \theta = 2\pi $:
   $
   Y = -\frac{1}{\pi} \cdot 2\pi + 1 = -1
   $

Thus, $ Y $ ranges from $ -1 $ to $ 1 $, i.e., $ Y \in [-1, 1] $.

### Step 2: Find the PDF of $ Y $ Using the Change of Variables Method

Since $ Y $ is a transformation of $\Theta$, we can find the PDF $ f_Y(y) $ by using the change of variables formula:
$
f_Y(y) = f_\Theta(\theta) \left| \frac{d\theta}{dy} \right|
$

#### Compute $ \frac{d\theta}{dy} $
From the equation $ \Theta = -\pi Y + \pi $, we differentiate with respect to $ y $:
$
\frac{d\theta}{dy} = -\pi
$
Thus, $ \left| \frac{d\theta}{dy} \right| = \pi $.

#### Substitute $ f_\Theta(\theta) $ and $\theta$ in Terms of $ y $
The PDF $ f_\Theta(\theta) $ is given by:
$
f_\Theta(\theta) = 
\begin{cases}
\frac{1}{2\pi} & 0 \leq \theta \leq 2\pi \\
0 & \text{otherwise}
\end{cases}
$

Since $ \theta = -\pi y + \pi $, this condition $ 0 \leq \theta \leq 2\pi $ translates to $ y \in [-1, 1] $. Therefore:
$
f_Y(y) = f_\Theta(-\pi y + \pi) \cdot \pi
$

For $ y \in [-1, 1] $, we have $ f_\Theta(-\pi y + \pi) = \frac{1}{2\pi} $. Substituting this in, we get:
$
f_Y(y) = \frac{1}{2\pi} \cdot \pi = \frac{1}{2}
$

### Final Answer

The PDF of $ Y $ is:
$
f_Y(y) = 
\begin{cases}
\frac{1}{2} & -1 \leq y \leq 1 \\
0 & \text{otherwise}
\end{cases}
$

This is a uniform distribution over the interval $[-1, 1]$.

```julia
using Plots

# Define the PDF of Y
function f_Y(y)
    if -1 <= y <= 1
        return 1 / 2
    else
        return 0
    end
end

# Generate a range of y values for plotting
y_vals = -1.5:0.01:1.5
f_Y_vals = [f_Y(y) for y in y_vals]

# Plot the PDF of Y
plot(y_vals, f_Y_vals, label=L"f_Y(y) = 1/2", xlabel="y", ylabel=L"f_Y(y)", 
     title="PDF of Y", linewidth=2, legend=:topright, ylim=(0, 0.6))
# display(plot!)
```

Example:
 
 Consider the transformation $y = x^2$. If $F\_X(x) = 0.5 exp(-|x|)$, find $F_y(y)$.

To find $ F_Y(y) $, the cumulative distribution function (CDF) of $ Y = X^2 $, given that $ F_X(x) = 0.5 e^{-|x|} $ is the CDF of $ X $, we can follow these steps.

### Step 1: Identify the PDF of $ X $

To find the CDF of $ Y $, we first need the probability density function (PDF) of $ X $, $ f_X(x) $. The CDF of $ X $ is given by:
$
F_X(x) = 0.5 e^{-|x|}
$

Since $ F_X(x) $ represents the probability that $ X \leq x $, we need to find $ f_X(x) = \frac{d}{dx} F_X(x) $.

#### Find $ f_X(x) $ by Differentiating $ F_X(x) $

1. **For $ x \geq 0 $:**
   $
   F_X(x) = 0.5 (1 - e^{-x})
   $
   Differentiating with respect to $ x $:
   $
   f_X(x) = \frac{d}{dx} \left(0.5 (1 - e^{-x})\right) = 0.5 e^{-x}, \quad x \geq 0
   $

2. **For $ x < 0 $:**
   $
   F_X(x) = 0.5 e^{x}
   $
   Differentiating with respect to $ x $:
   $
   f_X(x) = \frac{d}{dx} (0.5 e^x) = 0.5 e^{x}, \quad x < 0
   $

So, the PDF $ f_X(x) $ is:
$
f_X(x) = 
\begin{cases}
0.5 e^{-x} & x \geq 0 \\
0.5 e^{x} & x < 0
\end{cases}
$

### Step 2: Express $ F_Y(y) $ in Terms of $ X $

Since $ Y = X^2 $, we have $ y = x^2 $ and $ x = \pm \sqrt{y} $. Then:
$
F_Y(y) = P(Y \leq y) = P(X^2 \leq y)
$

This is equivalent to finding the probability that $ X $ lies within the interval $ [-\sqrt{y}, \sqrt{y}] $:
$
F_Y(y) = P(-\sqrt{y} \leq X \leq \sqrt{y})
$

Using the CDF $ F_X(x) $, we can write:
$
F_Y(y) = F_X(\sqrt{y}) - F_X(-\sqrt{y})
$

### Step 3: Substitute $ F_X(x) $ for $ \pm \sqrt{y} $

1. For $ F_X(\sqrt{y}) $ (where $ \sqrt{y} \geq 0 $):
   $
   F_X(\sqrt{y}) = 0.5 \left(1 - e^{-\sqrt{y}}\right)
   $

2. For $ F_X(-\sqrt{y}) $ (where $ -\sqrt{y} < 0 $):
   $
   F_X(-\sqrt{y}) = 0.5 e^{\sqrt{y}}
   $

Thus,
$
F_Y(y) = 0.5 \left(1 - e^{-\sqrt{y}}\right) - 0.5 e^{\sqrt{y}}
$

### Final Answer

The CDF $ F_Y(y) $ of $ Y = X^2 $ is:
$
F_Y(y) = 0.5 \left(1 - e^{-\sqrt{y}}\right) - 0.5 e^{\sqrt{y}}
$
for $ y \geq 0 $.

```julia
# Define the PDF of Y
function f_Y(y)
    if y > 0
        return (exp(-sqrt(y)) + exp(sqrt(y))) / (4 * sqrt(y))
    else
        return 0
    end
end

# Generate a range of y values for plotting
y_vals = 0.01:0.01:4 # Start slightly above 0 to avoid division by zero
f_Y_vals = [f_Y(y) for y in y_vals]

# Plot the PDF of Y
plot(y_vals, f_Y_vals, label=L"f_Y(y)", xlabel="y", ylabel=L"f_Y(y)", 
     title="PDF of Y", linewidth=2, legend=:topright)
```

Example: 

Consider the joint pdf

$$
f_{XY}(x,y) =
  \begin{cases}
    A e^-(2x+y)       & \quad , x, y \geq 0 \\
    0  & \quad \text{otherwise }
  \end{cases}

Noting that $X$ and $Y$ are statistically independent, determine the expectation of $g(X, Y) = X Y$.

To determine the expectation $ \mathbb{E}[g(X, Y)] = \mathbb{E}[X Y] $ given that $ X $ and $ Y $ are statistically independent, we can proceed as follows:

### Step 1: Find the Constant $ A $ for the Joint PDF

The joint PDF is given by:
$
f_{XY}(x, y) = 
\begin{cases} 
A e^{-(2x + y)} & x, y \geq 0 \\ 
0 & \text{otherwise}
\end{cases}
$
Since $ f_{XY}(x, y) $ is a valid joint probability density function, it must integrate to 1 over the support of $ x $ and $ y $. Therefore:
$
\int_0^{\infty} \int_0^{\infty} A e^{-(2x + y)} \, dx \, dy = 1
$

1. **Integrate with respect to $ x $:**
   $
   \int_0^{\infty} A e^{-(2x + y)} \, dx = A \int_0^{\infty} e^{-2x} e^{-y} \, dx = A e^{-y} \int_0^{\infty} e^{-2x} \, dx
   $
   The integral with respect to $ x $ is:
   $
   \int_0^{\infty} e^{-2x} \, dx = \left[ -\frac{1}{2} e^{-2x} \right]_0^{\infty} = \frac{1}{2}
   $
   So,
   $
   \int_0^{\infty} A e^{-(2x + y)} \, dx = A e^{-y} \cdot \frac{1}{2} = \frac{A}{2} e^{-y}
   $

2. **Integrate with respect to $ y $:**
   Now integrate with respect to $ y $:
   $
   \int_0^{\infty} \frac{A}{2} e^{-y} \, dy = \frac{A}{2} \int_0^{\infty} e^{-y} \, dy
   $
   The integral of $ e^{-y} $ is:
   $
   \int_0^{\infty} e^{-y} \, dy = \left[ -e^{-y} \right]_0^{\infty} = 1
   $
   Thus,
   $
   \int_0^{\infty} \int_0^{\infty} A e^{-(2x + y)} \, dx \, dy = \frac{A}{2} \cdot 1 = \frac{A}{2}
   $
   Setting this equal to 1, we find:
   $
   \frac{A}{2} = 1 \Rightarrow A = 2
   $

So, the joint PDF is:
$
f_{XY}(x, y) = 
\begin{cases} 
2 e^{-(2x + y)} & x, y \geq 0 \\ 
0 & \text{otherwise}
\end{cases}
$

### Step 2: Find $ \mathbb{E}[X Y] $ Using Independence

Since $ X $ and $ Y $ are independent, we have:
$
\mathbb{E}[X Y] = \mathbb{E}[X] \cdot \mathbb{E}[Y]
$

#### Find $ \mathbb{E}[X] $
To find $ \mathbb{E}[X] $, we need the marginal PDF of $ X $, $ f_X(x) $, by integrating $ f_{XY}(x, y) $ over $ y $:
$
f_X(x) = \int_0^{\infty} f_{XY}(x, y) \, dy = \int_0^{\infty} 2 e^{-(2x + y)} \, dy
$
1. Factor out terms involving $ x $:
   $
   f_X(x) = 2 e^{-2x} \int_0^{\infty} e^{-y} \, dy
   $
2. Evaluate the integral over $ y $:
   $
   \int_0^{\infty} e^{-y} \, dy = 1
   $
   Thus,
   $
   f_X(x) = 2 e^{-2x}, \quad x \geq 0
   $
   This is the PDF of an exponential distribution with parameter $ 2 $, so:
   $
   \mathbb{E}[X] = \frac{1}{2}
   $

#### Find $ \mathbb{E}[Y] $
Similarly, to find $ \mathbb{E}[Y] $, we need the marginal PDF of $ Y $, $ f_Y(y) $, by integrating $ f_{XY}(x, y) $ over $ x $:
$
f_Y(y) = \int_0^{\infty} f_{XY}(x, y) \, dx = \int_0^{\infty} 2 e^{-(2x + y)} \, dx
$
1. Factor out terms involving $ y $:
   $
   f_Y(y) = 2 e^{-y} \int_0^{\infty} e^{-2x} \, dx
   $
2. Evaluate the integral over $ x $:
   $
   \int_0^{\infty} e^{-2x} \, dx = \frac{1}{2}
   $
   Therefore,
   $
   f_Y(y) = e^{-y}, \quad y \geq 0
   $
   This is the PDF of an exponential distribution with parameter $ 1 $, so:
   $
   \mathbb{E}[Y] = 1
   $

### Step 3: Compute $ \mathbb{E}[X Y] $

Now we can compute $ \mathbb{E}[X Y] $ using independence:
$
\mathbb{E}[X Y] = \mathbb{E}[X] \cdot \mathbb{E}[Y] = \frac{1}{2} \cdot 1 = \frac{1}{2}
$

### Final Answer

The expectation $ \mathbb{E}[X Y] $ is:
$
\mathbb{E}[X Y] = \frac{1}{2}
$
