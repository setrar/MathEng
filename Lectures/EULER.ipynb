# Imaginary Numbers

## $1 + i = \sqrt{2} \cos 45^\circ + i \sqrt{2} \sin 45^\circ = \sqrt{2} \, e^{i (45^\circ)}$ 

### **Step 1: Understanding the Complex Number**
The number **$1 + i$** is a complex number, which means it has:  
- **Real part:** $1$  
- **Imaginary part:** $i$, which is equivalent to $1i$  

We can represent complex numbers in two common forms:  
1. **Rectangular form**: $x + iy$ (like $1 + i$)  
2. **Polar form**: $r (\cos \theta + i \sin \theta)$, where:
   - $r$ is the **magnitude** (or length)
   - $\theta$ is the **angle** (or direction)

### **Step 2: Find the Magnitude $r$**
The magnitude (also called modulus) of a complex number $x + iy$ is:

$r = |z| = \sqrt{x^2 + y^2}$

For $z = 1 + i$:

$r = \sqrt{x^2 + y^2} = \sqrt{1^2 + 1^2}$

$\boxed{ r = \sqrt{1 + 1} = \sqrt{2} }$

### **Step 3: Find the Angle $\theta$**
The angle $\theta$ (also called argument) is found using:
$\theta = \tan^{-1} \left(\frac{y}{x}\right)$

For $z = 1 + i$:

$\theta = \tan^{-1} \left(\frac{y}{x}\right) = \tan^{-1} \left(\frac{1}{1}\right)$

$\boxed{ \theta  = \tan^{-1} (1) = 45^\circ }$


### **Step 4: Express in Polar Form**
From complex number theory, we can express any complex number as:

$z = r (\cos \theta + i \sin \theta)$

Substituting $r = \sqrt{2}$ and $\theta = 45^\circ$:

$\boxed{ 1 + i = \sqrt{2} (\cos 45^\circ + i \sin 45^\circ) = \sqrt{2} \, e^{i (45^\circ)} }$

### **Final Answer**
This shows that $1 + i$ can be rewritten in polar form using the angle and magnitude.

## $1 + 1.5i = \sqrt{3.25} \left( \cos 56.3^\circ + i \sin 56.3^\circ \right)  = \sqrt{3.25} \, e^{i (56.3^\circ)}$

We want to express the complex number:

$1 + 1.5i$

in polar form, which involves finding its **magnitude** and **angle**.

### **Step 1: Compute the Magnitude $r$**
The magnitude of a complex number $z = a + bi$ is:

$r = |z| = \sqrt{a^2 + b^2}$

For $z = 1 + 1.5i$:

$r = \sqrt{1^2 + (1.5)^2} = \sqrt{1 + 2.25} = \sqrt{3.25}$

Approximating:

$\sqrt{3.25} \approx 1.803$

### **Step 2: Compute the Angle $\theta$**
The angle $\theta$ (argument of $z$) is given by:

$\theta = \tan^{-1} \left(\frac{b}{a}\right)$

For $z = 1 + 1.5i$:

$\theta = \tan^{-1} \left(\frac{1.5}{1}\right) = \tan^{-1} (1.5)$

Approximating:

$\tan^{-1} (1.5) \approx 56.3^\circ$

or in radians:

$\theta \approx 0.983 \text{ rad}$

### **Step 3: Express in Cosine Form**
The polar form of a complex number is:

$z = r (\cos \theta + i \sin \theta)$

Substituting the values:

$1 + 1.5i = 1.803 \left( \cos 56.3^\circ + i \sin 56.3^\circ \right)$

or in radians:

$1 + 1.5i = 1.803 \left( \cos 0.983 + i \sin 0.983 \right)$

### **Final Answer**
$\boxed{ 1 + 1.5i = \sqrt{3.25} \left( \cos 56.3^\circ + i \sin 56.3^\circ \right)  = \sqrt{3.25} \, e^{i (56.3^\circ)} }$

We start with the given equation:

$C \cdot \frac{dv}{dt} = i$

where the voltage is:

$v = A \cos ( \omega t + \theta )$

### **Step 1: Compute $\frac{dv}{dt}$**

Differentiating $v$ with respect to $t$:

$\frac{dv}{dt} = \frac{d}{dt} \left[ A \cos ( \omega t + \theta ) \right]$

Using the derivative of cosine:

$\frac{d}{dt} \cos(x) = -\sin(x)$

we get:

$\frac{dv}{dt} = -A \omega \sin ( \omega t + \theta )$

### **Step 2: Multiply by $C$**

$C \cdot \frac{dv}{dt} = C \left( - A \omega \sin ( \omega t + \theta ) \right)$

$i = -\omega C A \sin ( \omega t + \theta )$

$i = -\omega C A \sin ( \omega t + \theta )$
